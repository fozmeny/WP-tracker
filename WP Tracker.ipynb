```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
import datetime

# ChromeDriver yolunu belirtin
chromedriver_path = '/Users/daimon/Desktop/chromedriver-mac/chromedriver'

# Service nesnesi oluşturun
service = Service(executable_path=chromedriver_path)

# Chrome tarayıcı seçeneklerini ayarlayın
options = webdriver.ChromeOptions()
options.add_argument("--user-data-dir=chrome-data")  # WhatsApp oturumunun açık kalması için kullanıcı verilerini sakla

# WebDriver'ı başlatın
driver = webdriver.Chrome(service=service, options=options)

# WhatsApp Web'e gidin
driver.get('https://web.whatsapp.com')

# Kullanıcıyı takip etmek için telefon numarasını bulun
contact_name = 'Karımm❤️'

# Tarayıcıyı başlatmak ve QR kodunu taramak için kullanıcıya zaman verin
print("Lütfen QR kodunu tarayarak WhatsApp Web'e giriş yapın.")
time.sleep(30)

def check_online_status():
    try:
        # Sohbet listesindeki kullanıcıyı bulun ve tıklayın
        contact = driver.find_element(By.XPATH, f"//span[@title='{contact_name}']")
        contact.click()
        time.sleep(3)

        # Çevrimiçi durumunu kontrol edin
        status = driver.find_element(By.XPATH, "//span[@title='online']")
        return True
    except Exception as e:
        return False

try:
    while True:
        now = datetime.datetime.now()
        if check_online_status():
            print(f"{contact_name} is online at {now}")
        else:
            print(f"{contact_name} is offline at {now}")
        time.sleep(5)  # 5 saniyede bir kontrol edin
except KeyboardInterrupt:
    print("Takip durduruldu.")
finally:
    driver.quit()
```
